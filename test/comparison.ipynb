```python
def check_diff(x):
    temp = [re.sub('\(.*?\)','',a).strip() for a in x['cdn']]
    if sorted(x['cdn_x']) == sorted(x['cdn_y']) and sorted(x['cdn_x']) == sorted(temp):
        return True
    else:
        return False
```

```python
def check_diff_cf_ao(x):
    if sorted(x['cdn_x']) == sorted(x['cdn_y']):
        return True
    else:
        return False
```

```python
def check_diff_ao_mf(x):
    temp = [re.sub('\(.*?\)','',a).strip() for a in x['cdn']]
    if sorted(x['cdn_y']) == sorted(temp):
        return True
    else:
        return False
```

```python
def get_analysis(df):
    x = len(df[df['cdn_x']==df['groundtruth']]) / 10000
    y = len(df[df['cdn_y']==df['groundtruth']]) / 10000
    z = len(df[df['cdn']==df['groundtruth']]) / 10000
    print(f"acc: {x},{y},{z}")
    complete = 10000 - df['groundtruth'].value_counts().values[0]
    x1 = (len(df[df['cdn_x']==df['groundtruth']]) - df[df['cdn_x']==df['groundtruth']]['cdn_x'].value_counts().values[0]) / complete 
    y1 = (len(df[df['cdn_y']==df['groundtruth']]) - df[df['cdn_y']==df['groundtruth']]['cdn_y'].value_counts().values[0]) / complete
    z1 = (len(df[df['cdn']==df['groundtruth']]) - df[df['cdn']==df['groundtruth']]['cdn'].value_counts().values[0]) / complete
    print(f"com: {x1},{y1},{z1}")
```
